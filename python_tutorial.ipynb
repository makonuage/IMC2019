## 数値と文字列

```python
print(5%2)
print(5**2)
```

## 条件分岐と繰り返し ->リスト殺す

pythonの条件分岐は `if文` を用いて行なうことができます．
pythonは他の言語とは異なり **:(コロン)** と **インデント** を用いてブロックを表現しています．

```python
if 5>3:
    print('５は３よりも大きい')
else:
    print('５は３よりも小さい！？')
```

pythonの繰り返しには `for` を用いる方法と　`while` を用いる方法があります．  
`for` を用いた繰り返しはc言語の `for文` とは異なる記法を用います．

pythonでは `for 変数 in リスト` の形式を取っており，`リスト` の要素が１ループごとに `変数` に格納されて実行されます．

```python
for i in [1,2,3,4,5]:
    print(i)
```

しかし，いちいちリストを手で記述するのは面倒であるため以下のような書き方が用意されている．

```python
for i in range(30):
    print(i)
```

一方 `while文` はおおよそc言語と同じ構文であり，条件が真であるかぎりブロック内の処理を繰り返します．  

```python
hoge = 2
while hoge < 10000:
    print(hoge)
    hoge *= 2
```

pythonには**リスト内包表記**という記法があります．  
この記法を用いることで自由にリストを作成することができます．

```python
sugoi_list = [i*2 for i in range(10)]
print(sugoi_list)
```

### 例題(条件分岐と繰り返し)
以下の例題をやってみましょう.
1. hoge
2. hoge
3. hoge
4. hoge
5. hoge

```python
def hoge(x, lim):
    count = 1
    while x < lim:
        x *= 2
        count += 1
    return count-1, x//2

print(hoge(2, 1000))
```
