### 行列

>Pythonには標準で行列を扱う機能がありません（もちろん自力で実装はできるが、非常に困難）。
行列計算は多くの人が使う機能であるため、有志が作成した**numpy**という行列をはじめとした数値計算のための外部ライブラリ（pythonに付属していないが、追加で導入できる機能群）があります。

>numpyを使ったプログラム例を示します。
このプログラムは$2 \times 2$行列$A, B$の内積$C$を計算します。
記号を大文字にしたのは趣味です。

````python
import numpy

A = numpy.array([[1, 2], [3, 4]])
B = numpy.array([[5, 6],
                 [7, 8]])
print(A)
print(B)

C = A @ B
print(C)
````

>1行目でnumpyを読み込み、以降のプログラムで使えるようにします。
3, 4行目で行列A, Bを定義します。
numpyの行列なので、`array`の前に`numpy`を`.`で付けています。
9行目で内積を計算します。
`@`は内積の演算子（計算記号）です。

>行列同士の演算は内積以外にも、

>- 要素和
>- 要素差
>- 要素積
>- 要素商

>などが定義されています。

````python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Aの転置:")
print(A.T)

print("要素和:")
print(A + B)

print("要素差:")
print(A - B)

print("要素積:")
print(A * B)

print("要素商:")
print(A / B)
````

>このプログラムでは別の方法でnumpyをインポートしています。
`import`でnumpyを読み込むのは同様ですが、変数`np`にnumpyを割り当てています。
こうするとnumpyの代わりに`np`を接頭語として使うことができ、指の関節を節約できます。

>行列の`i`列`j`行要素を取り出すには次のようにします。インデックスが0スタートであることに気をつけましょう。

````python
import numpy as np
import numpy.linalg as nl

A = np.array([[1, 2], [3, 4]])
print(A)

print("1, 1要素")
print(A[0, 0])

print("2, 1要素")
print(A[1, 0])
````

>また,逆行列,行列式の計算は以下のように行います.
逆行列、行列式はnumpyだけでは計算できないので、さらに`numpy.linalg`をインポートしています。

````python
import numpy as np
import numpy.linalg as nl

A = np.array([[1, 2], [3, 4]])

print("逆行列:")
print(nl.inv(A))

print("行列式:")
print(nl.det(A))
````

## 関数

>関数は一般的にあるデータを受け取り，何らかの規則で変換した結果を返すものです。
Pythonで関数を作る方法は`def`を使う方法と`lambda`を使う方法がありますが、ここでは`lambda`を使った無名関数を作る方法を説明します.
無名関数は以下で定義できます.

>`lambda 引数:戻り値(処理の内容)`

>例えば,次のセルでは受け取った値$x$の平方$\sqrt{x}$を返す関数`sqrt`を定義して、$\sqrt{2}$を計算しています。

````python
# 目的: xの平方根を計算する
# 型: float -> float
sqrt = lambda x:x**0.5

print(sqrt(2))
````

>関数を定義することの利点は可読性を維持しつつ処理を繰り返すことができる点です。
処理を関数宣言の1箇所にまとめてほかの箇所で使いまわすことで、修正が必要になったとき変更すべき箇所が減ることも利点です。
これらの利点はプログラムの保守性を高めます。

>関数を使いまわす例を見てみましょう。
次の`mid_point`はベクトル（`np.ndarray`型）で与えられた2点の中点を求める関数です。

>上の関数では引数xを0.5乗したものを戻り値としています.

>条件分岐や繰り返しを含む複雑な関数の作り方については後で解説します.

````python
# 目的: 与えられたベクトルp1, p2の中点を計算する
# 型: (np.ndarray, np.ndarray) -> np.ndarray
mid_point = lambda p1, p2: (p1 + p2) / 2
````

>2点間を1:3で内分する処理は次のように`mid_point`関数を再利用して構築できます。

````python
# 目的: 与えられた2点間を1:3に内分する点を計算する
# 型: (np.ndarray, np.ndarray) -> np.ndarray
divide_1_3 = lambda p1, p2: mid_point(p1, mid_point(p1, p2))
````

````python
point_left = np.array([0, 0])
point_right = np.array([4, 8])

point_m = mid_point(point_left, point_right)
point_q = divide_1_3(point_left, point_right)

print(point_m)
print(point_q)
````

>プログラムを書くときのコツとして、**関数の入力と出力が何であるか**を意識して書く方法があります。
関数を書くときは上の例のように、何を受け取り何を求めるかを示す**目的**と、受け取るデータと返すデータの**型**をコメント文で明記すべきです。

>コメントを書くのは多少めんどうですが、習熟速度が段違いに変わるので**必ず**書くようにしましょう。

## 条件分岐

>ある条件を満たすときだけ処理を実行する、または満たさないとき処理を実行するためにpythonではif文やif式を使うことができます。

>条件とは真偽型の値を返す式や関数、あるいは真偽値そのものを指します。
例えば「値`x`が偶数である」という条件は剰余を求める演算子`%`と両辺が等しいかを判定する演算子`==`を使って次のように書くことができます。

````python
x = 42
condition = x % 2 == 0

print(condition)
````

>値`42`は偶数なので、`condition`の値は`True`になっています。
奇数の場合も示します。

````python
x = 43

print(x % 2 == 0)
````

>こちらでは変数に代入しないで条件式の値を直接プリントしました。
この場合は`x`が偶数でないので、`False`が出力されました。

>条件判定に使う`==`などの演算子はまとめて条件演算子と呼ばれます。
よく使うほかの条件演算子を次に示します。

| 記号 | 意味          |
| ---- | ------------- |
| ==   | 等号 $=$      |
| !=   | 等号 $\neq$   |
| <    | 不等号 $<$    |
| >    | 不等号 $>$    |
| <=   | 不等号 $\leq$ |
| >=   | 不等号 $\geq$ |

>これらの演算子を使って処理を切り替える方法を見ていきます。

### if文

>if文は次の形式で書きます。

>```
>if 条件式1:
>    条件1を満たすときの処理
>elif 条件式2:
>    条件2を満たすときの処理
>else:
>    条件1, 2のどちらも満たさないときの処理
>```

>それぞれの処理は字下げ幅（先頭の空白の幅）を揃えることで複数列挙できます。
`elif`, `else`は必要なければ省略できます。
if文は複数行を使って書かなければならないため、原則1行で書かなければならない無名関数の中で使うことはできません。

>次のコードは`year`がうるう年かどうかに応じて出力する文字を切り替えます。
if文が二重になっているため少し読みづらいですが、`year`の値を色々変えながら処理の流れを探ってみてください。

````python
year = 2019

if year % 400 == 0: # 西暦年が400で割り切れる年は必ずうるう年
    print(f"{year}年はうるう年")
elif year % 4 == 0: # 西暦年が4で割り切れる年は（原則として）うるう年
    if year % 100 == 0: # 西暦年が100で割り切れる年は（原則として）平年
        print(f"{year}年は平年")
    else:
        print(f"{year}年はうるう年")
else:
    print(f"{year}年は平年")
````

### if式

>if式は次の形式で書きます。

>```
>値1 if 条件式 else 値2
>```

>条件式を満たすときは値1、満たさないときは値2が使われます。
if文と違って`elif`はありませんし、`else`を省略することができません。
これは必ずなんらかの値を返すための制約です。
その代わりif式はその他の計算式に埋め込むことができ、すなわち無名関数の中で使うことができます。

>if式を使うとクロネッカーのデルタ、

>\begin{equation}
  \delta_{i, j}
  =
  \begin{cases}
    1 & i = j \\
    0 & \text{otherwise}
  \end{cases}
\end{equation}

>は次の無名関数として実装できます。

````python
# 目的: クロネッカーのデルタ（iとjが等しいとき1、それ以外は0を返す）を実装する
# 型: (int, int) -> int
kronecker_delta = lambda i, j: 1 if i == j else 0

print(kronecker_delta(1, 2))
print(kronecker_delta(1, 1))
````

### for文

>Pythonのfor文はC言語のfor文とは異なる動作をします。
Pythonのforはイテレータ（`np.ndarray`型やここでは説明していないリスト型など）の各要素に対して処理を行うときに使われます。

>整数`0`から`9`までを順番に画面表示するコードを示します。

````python
import numpy as np

for i in np.arange(10):
    print(i)
````

>ここで`np.arange(10)`は`np.ndarray([0, 1, ..., 9])`を返す関数です。
つまりこの処理は`np.ndarray`の要素である`0`, `1`, ..., `9`を順番に変数`i`に束縛し、それぞれの`i`に関して`print`関数を呼び出しています。
>このイテレータは連番である必要はなく、例えば次のようなベクトル`v`にもfor文を適用できます。

>このイテレータは連番である必要はなく、例えば次のようなベクトル`v`にもfor文を適用できます。

````python
import numpy as np

v = np.array([3.14, 42.0, 1701.0])

for f in v:
    print(f * 10.0)
````

>for文は二重、三重にできます。

````python
for i in np.arange(3):
    for j in np.arange(3):
        print(i + j)
````

>ここで気をつけてほしいのが、`v`は**行列**ではなく**ベクトル**である点です。
違いをはっきりさせるために、意味的には等価ですが次元が異なる行列`M`とベクトル`m`を示します。

````python
M = np.array([[1, 2, 3]])
m = np.array([1, 2, 3])

print("行列M")
print(M)
print("ベクトルm")
print(m)
````

>`M`は2次元の配列として定義された行列ですが、`m`は1次元の配列として定義されたベクトルです。
このことはそれぞれに`np.ndim`関数を適用した結果からも確認できます。

````python
print("行列Mの次数")
print(np.ndim(M))
print("ベクトルmの次数")
print(np.ndim(m))
````

>行列とベクトルにfor文を適用すると異なる結果になります。

````python
print("行列Mの場合")
for i in M:
    print(i)
print("ベクトルmの場合")
for i in m:
    print(i)
````

>行列`M`にfor文を適用すると、1行目`[1 2 3]`が丸ごと取り出されます。
`M`に対する動作を`m`と同じようにするには、`np.reshape`関数で`M`を1次元にしてから使いましょう。

````python
# (1, 3)行列を(3,)ベクトルに変換する。
# 数字数えるのがめんどうなら、
# (-1,)と書くと自動でやってくれる。
reshaped_M = np.reshape(M, (3,))

for i in reshaped_M:
    print(i)
````

### while文

>while文の働きはC言語のものとよく似ています。
括弧がない点、コロンを使う点を除くと文法も同じなので、ここでは使用例だけ示します。
for文の練習問題1はwhile文を使うと次のように書けます。

````python
total = 0
i = 1
while i <= 100:
    total = total + i
    i = i + 1

print(total)
````

>ただしfor文を使うほうがスッキリするので普通はやりません。

>while文を使うのは終了条件を満たすまで処理を繰り返したいときです。
次の例は[ニュートン法](https://ja.wikipedia.org/wiki/%E3%83%8B%E3%83%A5%E3%83%BC%E3%83%88%E3%83%B3%E6%B3%95)というアルゴリズムを使って方程式$x^2 - 42 = 0$の解、すなわち$\sqrt{42}$を計算しています。
解の更新量`update`がしきい値$1 \times 10^{-10}$を下回るまで更新を繰り返します。

````python
import numpy as np

f = lambda x: x**2 - 42
df = lambda x: 2*x

x = 3.14
update = f(x)/df(x)

while abs(update) > 1E-10:
    x = x - update
    update = f(x)/df(x)

print(x)
````

### グラフ作成モジュール(matplotlib)

>matplotlibはグラフを表示するためのライブラリです。
実用上はmatplotlibのサブモジュールであるmatplotlib.pyplotを使うことになります。

>`plot`関数でグラフを描画し、`show`関数で画面に表示します。
`show`するまでは`plot`したものは確認できないので、`show`を呼ぶことを忘れないようにしましょう。

````python
import numpy as np
import matplotlib.pyplot as plt

# 0~2piの範囲のsin(x)を計算する
sine = 12*np.sin(np.linspace(0, 2*np.pi, 201))
cose = 9*np.cos(np.linspace(0, 2*np.pi, 201))

# `sine`を線グラフにする
plt.plot(sine)
plt.plot(cose)

# グラフを表示する
plt.show()
````

>`plot`の引数に`label=`を与えることで、線に名前をつけれます。
`show`の前に`legend`を忘れないようにしましょう。

````python
import numpy as np
import matplotlib.pyplot as plt

# 横軸
time = np.linspace(0, 2*np.pi, 201)

# 0~2piの範囲のsin(x)を計算する
sine = 12*np.sin(time)
cose = 9*np.cos(time)

# `sine`を線グラフにする
plt.plot(time, sine, label="sine") # ラベルつき
plt.plot(time, cose, label="cose") # ラベルつき

plt.legend() # ラベルをグラフ内に表示

# グラフを表示する
plt.show()
````

>軸ラベルは`xlabel`, `ylabel`でつけれます。

````python
import numpy as np
import matplotlib.pyplot as plt

# 横軸
time = np.linspace(0, 2*np.pi, 201)

# 0~2piの範囲のsin(x)を計算する
sine = 12*np.sin(time)
cose = 9*np.cos(time)

# `sine`を線グラフにする
plt.plot(time, sine)
plt.plot(time, cose)

plt.xlabel("Time")
plt.ylabel("Value")

# グラフを表示する
plt.show()
````
